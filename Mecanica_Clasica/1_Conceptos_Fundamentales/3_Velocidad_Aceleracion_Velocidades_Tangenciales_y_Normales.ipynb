# Velocidad, Aceleración, Velocidades Tangenciales y Normales.

## Velocidad

### Velocidad

la velocidad se define como el cambio de posición de un objeto respecto al tiempo.

Dado esto se define como la variación de la posición respecto a un intervalo de tiempo:

$$
\vec{v} = \frac{\vec{\triangle r}}{\triangle t}
$$

También se puede tomar la trayectoria dentro de un intervalo del tiempo:

$$
v = \frac{\triangle s}{\triangle t}
$$

### Velocidad instantánea

La velocidad instantánea es la velocidad dentro de un instante de tiempo, La velocidad instantánea es siempre tangente a la trayectoria.

$$
\vec{v} = \lim_{\triangle t \to 0} \frac{\vec{\triangle r}}{\triangle t} = \frac{d \vec{r}}{d t}
$$

- $ \vec{v} $: Vector velocidad.
- $ \vec{\triangle r} $: Vector incremento de la posición ($ \vec{\triangle r} = \vec{r_f} - \vec{r_o} $).
  - $ \vec{r_f} $: Vector posición final. 
  - $ \vec{r_o} $: Vector posición inicial. 
- $ \triangle s $: Valor escalar del incremento de la trayectoria ($ \triangle s = s_f - s_o $).
  - $ s_f $: trayectoria final.
  - $ s_o $: trayectoria inicial.
- $ \triangle t $: Intervalo de tiempo ($ \triangle t = t_f - t_o $).
  - $ t_f $: Tiempo final.
  - $ t_o $: Tiempo inicial.


## Velocidad angular

Se define con un cuerpo con una trayectoria de movimiento en rotación.

![vectores angulares](./img/vectores_angulares.png)

### Velocidad angular

$$
\omega = \frac{\triangle \phi}{\triangle t}
$$

### Velocidad angular instantánea

$$
\omega = \lim_{\triangle t \to 0} \frac{\triangle \phi}{\triangle t} = \frac{d \phi}{d t}
$$

### Vector velocidad angular

El vector velocidad angular ($ \vec{\omega} $) es un vector axial paralelo al eje de rotación, cuyo modulo es la velocidad angular definida anteriormente ($ \omega $).

![vector velocidad angular](./img/vector_velocidad_angular.png)

$$
\vec{\omega} = \frac{d \phi}{d t} \vec{e} => \omega \vec{e} = \frac{d \phi}{d t}
\\
\vec{e_t} = \vec{e_n} \times \vec{e}
$$

### Velocidad tangencial

$$
v = \omega \cdot r, ~~~~ \vec{v} = v e_t = r \omega (\vec{e_n} \times \vec{e}) = (r \vec{e_n}) \times (w \vec{e}) = \vec{PO} \times \vec{\omega} = \vec{\omega} \times \vec{OP} 
$$

- $ v $: Velocidad tangencial
- $ \omega $: Velocidad angular
- $ \triangle \phi $: Incremento de la trayectoria circular ($ \phi_f - \phi_o $).
  - $ \phi_f $: Final de la trayectoria.
  - $ \phi_o $: Inicio de la trayectoria.
- $ \triangle t $: Intervalo de tiempo ($ \triangle t = t_f - t_o $).
  - $ t_f $: Tiempo final.
  - $ t_o $: Tiempo inicial.
- $ \vec{\omega} $: Vector velocidad angular.
- $ \vec{e} $: Vector (unitario) que indica la dirección del eje.
- $ \vec{e_t} $: Vector (unitario) tangencial.
- $ \vec{e_n} $: Vector (unitario) normal.
- $ r $: Radio de giro del movimiento.
- $ \vec{OP} $: Vector posición del punto $ P $.

## Aceleración

### Aceleración

La aceleración se define como el cambio de velocidad dentro de un instante de tiempo.

$$
\vec{a} = \frac{\vec{\triangle v}}{\triangle t}
$$

## Aceleración instantánea

Se define como la aceleración dada dentro de un instante de tiempo.

$$
\vec{a} = \lim_{\triangle t \to 0} \frac{\vec{\triangle v}}{\triangle t} = \frac{d \vec{v}}{d t}
\\
\vec{a} = \frac{d^2 \vec{r}}{d t^2}
$$

- $ \vec{a} $: Aceleración.
- $ \vec{\triangle v} $: Incremento de la velocidad ($ \vec{v_f} - \vec{v_o} $).
  - $ \vec{v_f} $: Velocidad final.
  - $ \vec{v_o} $: Velocidad inicial.
- $ \triangle t $: Intervalo de tiempo ($ \triangle t = t_f - t_o $).
  - $ t_f $: Tiempo final.
  - $ t_o $: Tiempo inicial.
- $ \vec{r} $: Vector Posición.

## Aceleración angular

Se define con un cuerpo con una trayectoria de movimiento en rotación con velocidad variable.

![aceleraciones angulares](./img/aceleraciones_angulares.png)

### Aceleración angular

$$
\alpha = \frac{\triangle \omega}{\triangle t}
\\
\vec{\alpha} = \frac{\vec{\triangle \omega}}{\triangle t}
$$

### Aceleración angular instantánea

$$
\alpha = \lim_{\triangle t \to 0} \frac{\triangle \omega}{\triangle t} = \frac{d \omega}{d t}
\\
\alpha = \frac{d^2 \phi}{d t^2}
$$

### Aceleración tangencial

$$
a_t = \frac{d v}{d t}
\\
\vec{a} = \frac{d \vec{v}}{d t} = \frac{d}{d t} (v \cdot \vec{e_t})
\\
\frac{d}{d t} (v \cdot \vec{e_t}) = \frac{d v}{d t} \cdot \vec{e_t} + v \cdot \boxed{\frac{d \vec{e_t}}{d t}} = a_t \cdot \vec{e_t} + v \cdot \boxed{(\vec{\omega} \times \vec{e_t})}
$$

Definimos que es $ \frac{d \vec{e_t}}{d t} $:

$$
\vec{e_t} = \cos{\theta}\vec{i} + \sin{\theta}\vec{j}
\\
\frac{d \vec{e_t}}{d \theta} = -\sin{\theta}\vec{i} + \cos{\theta}\vec{j} = \vec{e_n}
\\
\frac{d \vec{e_t}}{d t} = \frac{d \vec{e_t}}{d \theta} \times \frac{d \theta}{d t} = \vec{e_n} \times \frac{d \theta}{d t}
$$

Dado que $ \frac{d \theta}{d t} = \omega $ y $ \omega = \frac{v}{r} $:

$$
\frac{d \vec{e_t}}{d t} = \vec{e_n} \times \vec{\omega} \times \vec{e} = \vec{e_n} \times (- \vec{e} \times \vec{\omega}) = - (\vec{e_n} \times \vec{e}) \times \vec{\omega} = - \vec{e_t} \times \vec{\omega} = \vec{\omega} \times \vec{e_t}
\\
\frac{d \vec{e_t}}{d t} = \vec{e_n} \cdot \frac{v}{r} 
$$

Dado  que la aceleración es:

$$
\vec{a} = \vec{a_t} + \vec{a_n}
\\
\vec{a} = a_t \cdot \vec{e_t} + v \cdot (\vec{\omega} \times \vec{e_t})
\\
==>
\\
\vec{a_t} = a_t \cdot \vec{e_t}
$$

### Aceleración centrifuga o normal

$$
\vec{a_c} = \vec{a_n} = v (\vec{\omega} \times \vec{e_t}) = \frac{v^2}{r} \cdot \vec{e_n}
$$
