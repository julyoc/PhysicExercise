```python
import numpy as np
import matplotlib.pyplot as plt
```

# Algebra Vectorial

Un vector es un objeto matem√°tico que contiene `N` valores de componentes.

$$
\vec{v} = (
    \begin{matrix}
        v_1 & v_2 & v_3 
    \end{matrix}
) = v_1 \vec{i} + v_2 \vec{j} + v_3 \vec{k}
$$

## Suma de vectores

Para la suma de vectores se suma cada una de las componentes del vector respectivamente.

$
\vec{v} + \vec{u} = (
    \begin{matrix}
        v_1 + u_1 & v_2 + u_2 & v_3 + u_3
    \end{matrix}
) = (v_1 + u_1) \vec{i} + (v_2 + u_2) \vec{j} + (v_3 + u_3) \vec{k}
$

### Propiedades de la suma de vectores

- Asociativa: $ \vec{v} + ( \vec{u} + \vec{w} ) = ( \vec{u} + \vec{v} ) + \vec{w} $
- Conmutativa: $ \vec{v} + \vec{u} = \vec{u} + \vec{v} $
- Elemento opuesto: $ \vec{v} + (- \vec{v}) = \vec{0} $
- Elemento neutro: $ \vec{v} + \vec{0} = \vec{v} $

```python
#https://danielmuellerkomorowska.com/2020/06/19/plotting-2d-vectors-with-matplotlib/
V = np.array([1, 5])
print('V = ', V)
U = np.array([4, 1])
print('U = ', U)
W = V + U
print('W = ', W)
O = np.array([[0, 0, 0], [0, 0, 0]])
print("O = ", O)
vectors = np.array((V, U, W))
print('vectors = ', vectors)

plt.quiver(*O, vectors[:,0], vectors[:,1], scale=1, scale_units='xy', angles='xy', color=['g', 'r', 'b'], dashes='-')
plt.quiver(vectors[1::-1,0], vectors[1::-1,1], vectors[:2,0], vectors[:2,1], scale=1, scale_units='xy', angles='xy', color=['g', 'r'], hatch='O')
plt.text(0, 3, 'V', color='green')
plt.text(3, 0, 'U', color='red')
plt.text(2.50, 4, 'V + U', color='blue')
plt.text(3, 6, 'U\'', color='red')
plt.text(4.6, 3, 'V\'', color='green')
plt.xlim([-1, 7])
plt.ylim([-1, 7])
```

## Producto escalar

Al realizar el producto de dos vectores devuelve un escalar como resultado.

$$
\vec{v} \cdot \vec{u} = v_1 \times u_1 + v_2 \times u_2 + v_3 \times u_3
\\
\vec{v} \cdot \vec{u} = |\vec{v}| |\vec{u}| \cos \alpha
$$

De forma matricial se representa como

$$
v \cdot u = v^T u
$$

### Propiedades

- $ ( \vec{v} + \vec{u} ) \cdot \vec{w} = \vec{v} \cdot \vec{w} + \vec{u} \cdot \vec{w} $
- $ ( \alpha \vec{v} + \beta \vec{u} ) \cdot \vec{w} = \alpha ( \vec{v} \cdot \vec{w}) + \beta ( \vec{u} \cdot \vec{w} ) $
- $ \vec{v} \cdot \vec{u} = \vec{u} \cdot \vec{v} $
- $ \vec{v} \cdot \vec{v} \geq 0 $
- $ \vec{v} \cdot \vec{v} = ||\vec{v}||^2 $
- Desigualdad de Cauchy-Bunyakovsky-Schwarz: $ \vec{v} \cdot \vec{u} \leq ||\vec{v}||\cdot||\vec{u}|| $
- Son ortogonales si: $ \vec{v} \cdot \vec{u} = 0 $

```python
O = np.array([[0, 0], [0, 0]])
vectors = np.array((V, U))
print('V = ', V)
print('U = ', U)
print('U * V = ', U.T.dot(V))
plt.quiver(*O, vectors[:,0], vectors[:,1], scale=1, scale_units='xy', angles='xy', color=['g', 'r', 'b'], dashes='-')
theta = np.linspace(0, np.pi / 2, 100)
cirX = np.cos(theta) + 0.3
cirY = np.sin(theta) + 0.35
plt.plot(cirX, cirY, color="grey")
plt.text(0, 3, 'V', color='green')
plt.text(3, 0, 'U', color='red')
#https://stackoverflow.com/questions/63514923/how-to-include-symbols-in-matplotlib-title
plt.text(1.25, 1, '$\\alpha$', color='grey')
plt.xlim([-1, 7])
plt.ylim([-1, 7])
```

## Producto Vectorial

El producto vectorial de un vector $ \vec{v} $ por otro $ \vec{u} $ nos da por resultado otro vector el cual es perpendicular al plano formado por los vectores $ \vec{v} $ y $ \vec{u} $.

$$
\vec{v} \times \vec{u} = \vec{r}
\\
|\vec{v} \times \vec{u}| = |\vec{v}| |\vec{u}| \sin{\alpha}
\\
% https://www.physicsread.com/latex-matrix-determinant/
\vec{v} \times \vec{u} = \begin{vmatrix} 
\vec{i} & \vec{j} & \vec{k} \\
v_1 & v_2 & v_3 \\
u_1 & u_2 & u_3
\end{vmatrix} = (v_2 \cdot u_3 - u_2 \cdot v_3) \vec{i} + (v_3 \cdot u_1 - u_3 \cdot v_1) \vec{j} + (v_1 \cdot u_2 - u_1 \cdot v_2) \vec{k} 
$$

- $ \vec{v} \times \vec{u} \neq \vec{u} \times \vec{v} => \vec{v} \times \vec{u} = \vec{w}, \vec{u} \times \vec{v} = - \vec{w} $
- $ \vec{v} \times (\vec{u} + \vec{w}) = \vec{v} \times \vec{u} + \vec{v} \times \vec{w} $

```python
# https://stackoverflow.com/questions/27023068/plotting-3d-vectors-using-python-matplotlib
from mpl_toolkits.mplot3d import Axes3D
```

```python
V = np.array([1, 2, 1])
print('V = ', V)
U = np.array([1, 2, -2])
print('U = ', U)
W = np.cross(V, U)
print('V x U = ', W)
O = np.array([0, 0, 0])
vectors = np.array((V, U, W))
vectors = np.array((np.append(O, V), np.append(O, U), np.append(O, W)))
print('vectors = ', vectors)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
x1, x2, x3, x4, x5, x6 = zip(*vectors)
ax.quiver(x1, x2, x3, x4, x5, x6, color=['r', 'g', 'b'])
ax.set_xlim([-3, 5])
ax.set_ylim([-3, 5])
ax.set_zlim([-2, 5])
ax.text(*V, 'V')
ax.text(*U, 'U')
ax.text(*W, 'W')
```
